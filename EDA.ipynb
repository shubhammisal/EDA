```python
data["Date"]=pd.to_datetime(data["Date"])
plt.figure(figsize=(20,5))
sb.lineplot(data=data, x="Date", y="Volume", color="dodgerblue")
plt.title('Volume over Date')
plt.show()
```

```python
plt.figure(figsize=(25,10))
sb.lineplot(data=data, x="Date", y="High", color="blue",label="High Price")
sb.lineplot(data=data, x="Date", y="Low", color="red",label="Low Price")
plt.title('High and Low Price over Date')
plt.legend(loc="upper left",fontsize=20)
plt.show()
```

```python
data["Year"] = data["Date"].dt.year
data["Month"] = data["Date"].dt.month
data["Day"] = data["Date"].dt.day

plt.figure(figsize=(25,10))
sb.lineplot(data=data, x="Year", y="Open",label="Open",color="yellow")
sb.lineplot(data=data, x="Year", y="Close",label="Close" ,color="red")
plt.title('Open vs Close USD over Year')
plt.legend(loc="upper left",fontsize=20)
plt.show()
```

```python
fig, ax = plt.subplots(6, 1, figsize = (15, 13))
sb.distplot(data['Open'], ax = ax[0])
sb.distplot(data['High'],ax = ax[1])
sb.distplot(data['Low'],ax = ax[2])
sb.distplot(data['Close'], ax = ax[3])
sb.distplot(data['Adj Close'],ax = ax[4])
sb.distplot(data['Volume'],ax = ax[5])
plt.tight_layout()
```
